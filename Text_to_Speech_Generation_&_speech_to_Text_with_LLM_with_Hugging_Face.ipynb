# **Import Libraries**

# **Text_to_Audio_Conversion_using_Hugging_face**

```python
from transformers import pipeline
```

# **Model**

```python
text ="Please Subscribe my channel: Content on Demand"

pipe=pipeline("text-to-speech", model='suno/bark-small')

output=pipe(text)
```

# **Vector Output**

```python
output
```

# **Text_To_Audio_Convert**

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
